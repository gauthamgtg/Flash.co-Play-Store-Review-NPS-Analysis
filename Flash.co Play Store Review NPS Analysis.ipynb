```python
from google_play_scraper import Sort,reviews_all

reviews = reviews_all(
                        'com.flashmonk.flash',
                        sleep_milliseconds=0,
                        sort=Sort.NEWEST)
```

```python
df = pd.DataFrame(np.array(reviews),columns=['review'])
df = df.join(pd.DataFrame(df.pop('review').tolist()))
df.head()
```
